```python
import re
import numpy as np
import pandas as pd
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

Parses line like this:

942/1000 [===========================>..] - ETA: 14s - loss: 2.6230 - regression_loss: 2.0926 - classification_loss: 0.5304

```python
pattern = '(\d+)/\d+ \[[.=>]+\] - ETA\: [\d:s]+ - loss: ([\d.e+]+) - regression_loss: ([\d.e+]+) - classification_loss: ([\d.e+]+)'
```

```python
def plot_losses(filepath):
    with open(filepath) as log:
        text = log.read()   
        
    rows = [[int(match[1]), float(match[2]), float(match[3]), float(match[4])] for match in re.finditer(pattern, text)]
    dataframe = pd.DataFrame(np.array(rows), columns=['step', 'loss', 'regression_loss', 'classification_loss'])
    
    fig, axis = plt.subplots(4, 1, figsize=(10, 20))
    dataframe.plot(ax=axis[0], y=['loss', 'regression_loss', 'classification_loss'])
    dataframe.plot(ax=axis[1], y='loss')
    dataframe.plot(ax=axis[2], y='regression_loss' ,c='DarkOrange')
    dataframe.plot(ax=axis[3], y='classification_loss', color='green')
    
    #print(dataframe.head(20))
    #print(dataframe.tail(20))
```

#### Training process for Mobilenetv3 backbone with batch_size=1 and lr=0.001

```python
plot_losses('mnv3_small_b1_001.log')
```

#### Training process for Mobilenetv3 backbone with batch_size=1 and lr=1e-5 (default)

```python
plot_losses('mnv3_small_b1.log')
```

#### Training process for Mobilenetv3 backbone with batch_size=1, lr=1e-5 and weighted losses

```python
plot_losses('mnv3_small_b1_wl.log')
```
